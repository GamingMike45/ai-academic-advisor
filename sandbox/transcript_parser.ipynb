```python
import json
import fitz
import re

def pdf_to_text(pdf_path):
    doc = fitz.open(pdf_path)
    transcript = ""
    # Iterate through each page and extract text
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        transcript += page.get_text() + "\n"
    doc.close()
    return transcript
```

```python
transcript1 = pdf_to_text('../raw_data/sample1.pdf')
print(transcript1)
```

```python
transcript2 = pdf_to_text('../raw_data/sample2.pdf')
print(transcript2)
```

```python
transcript3 = pdf_to_text('../raw_data/sample3.pdf')
transcript3
```

```python
sections = re.split(r'(?m)^(?=[A-Z][A-Z ]+$)', transcript3)
student_info_text = sections[1]

program = re.findall(r"(?m)^Program[\s|\t|:]*?\r?\n(.*)", student_info_text)
program = ", ".join([m.split(",")[0].strip() for m in program]) if program else None
print(program)
    
major = re.findall(r"(?m)^Major and Department[\s|\t|:]*?\r?\n(.*)", student_info_text)
major = ", ".join([m.split(",")[0].strip() for m in major]) if major else None
print(major)
```

```python
replacements = {
    "info" : "information",
    " ": "_",
    "&": "and",
}
pattern = re.compile("|".join(map(re.escape, replacements)))
filename_candidate = pattern.sub(lambda m: replacements[m.group(0)], "Info System & Analytics").lower()
print(filename_candidate)
```
