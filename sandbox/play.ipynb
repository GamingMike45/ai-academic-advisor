```python
import re
def extract_preq(desc):
    """
    Extracts the prerequisites from the course description
    """
    return desc.split('Prerequisite Courses:')[1]
    
def parse_preq(course, preq):
    course_pattern = r"[A-Z]{2,4} \d{5}"
    operator_pattern = r"\b(?:and|or)\b|\(|\)"
    out = re.findall(operator_pattern + '|' + course_pattern, preq)
    # joins the string and removes empty ()
    out = " ".join(out).replace("( )", "")
    # converts the preq boolean expression into valid python code 
    expr = re.sub(r"\b([A-Z]{2,4} \d{5})\b",r"('\1' in taken)", out)
    precomp = compile(expr, f"<{course}>", "eval")
    return out, expr, precomp
course = 'GEOG 16290' 
print(df[df['CourseCode'] == course].iloc[0]['Description'])
preq = extract_preq(df[df['CourseCode'] == course].iloc[0]['Description'])
out, expr, precomp = parse_preq(course, preq)
out
```

```python
print(re.findall(r'[A-Z]{2,4} \d{5}', out))
print(set(re.findall(r'[A-Z]{2,4} \d{5}', out)))
```

```python
# we can safely conclude that if none of the course are taken, then it would return false
# if all the courses are taken, then it would return true
# this only holds true, bc all the operators are only and/or

import itertools

# get all possible combinations of courses, create a set, to remove duplicates
s = set(re.findall(r'[A-Z]{2,4} \d{5}', out))
all_comb = []

for i in range(1, len(s)+1):
    all_comb.extend(itertools.combinations(s, i))
print(len(all_comb))
all_comb
```

```python
all_comb_output = []
for comb in all_comb:
    all_comb_output.append(eval(precomp, {"taken": comb}))

true_combs = [all_comb[i] for i, v in enumerate(all_comb_output) if v]
true_combs
```

```python
print(len(true_combs))
print(len(all_comb_output))
print(len(true_combs) / len(all_comb_output))
```
