## Department Abbreviation Extraction

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(open('departments_abbr.html'), 'html.parser')
print(soup.prettify())
```

```python
for department in soup.find_all('li'):
    print(department.text)
```

```python
REPLACEMENTS = {
    '&': 'and',
    "Eng.": 'Engineering',
    "Crit": "Critical",
    "Lit": "Literature",
    "Ed": "Education",
    "Plng": "Planning",
    "Sci": "Science",
    "Tech": "Technology",
    "Info": "Information",
    "Sys": "Systems",
    "Prof": "Professional",
    "Prgm": "Program",
    "Erly": "Early",
    "Elem": "Elementary",
    "Fnds": "Foundations",
    "Incl": "Inclusive",
    "Stdnt": "Student",
    "Persnl": "Personal",
    "Mgmt": "Management",
    "Engn": "Engineering",
}
```

```python
import re

departments = {}
for department in soup.find_all('li'):
    x = department.text.split('-')
    
    # replace any shortenings
    for old, new in REPLACEMENTS.items():
        pattern = re.compile(rf"\b{re.escape(old)}\b")
        x[1] = re.sub(pattern, new, x[1])
    
    departments[x[0]] = x[1]
    print(f"{x[0]}: {x[1]}")
```

```python
import json

# save departments to json
with open("departments_abbr.json", "w") as f:
    json.dump(departments, f)
```

## Subjects Abbreviation Extraction

```python
soup = BeautifulSoup(open('subjects_abbr.html'), 'html.parser')
print(soup.prettify())
```

```python
for subject in soup.find_all('li'):
    print(subject.text)
```

```python
REPLACEMENTS["Engr"] = "Engineering"
REPLACEMENTS["Dsgn"] = "Design"
REPLACEMENTS["Sculp"] = "Sculpture"
REPLACEMENTS["Transl"] = "Translational"
REPLACEMENTS["Educ"] = "Education"
REPLACEMENTS["Engnrg"] = "Engineering"
```

```python
REPLACEMENTS
```

```python
import re

subjects = {}

for subject in soup.find_all('li'):
    x = subject.text.split('-')
    
    # there are some subjects that don't have an abbreviation
    if len(x) != 2:
        subjects[x[0]] = x[0]
        continue
    
    for old, new in REPLACEMENTS.items():
        pattern = re.compile(rf"\b{re.escape(old)}\b")
        x[1] = re.sub(pattern, new, x[1])
    
    subjects[x[0]] = x[1]
```

```python
subjects
```

```python
import json

# save departments to json
with open("subjects_abbr.json", "w") as f:
    json.dump(subjects, f)
```
